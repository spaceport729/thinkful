```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
gender_df = pd.read_csv('UnmetFamPlanning_GenderInequality2.csv')
```

```python
plt.figure(figsize=(100,100))
```

```python
plt.subplot(2,2,1)
plt.scatter(x = gender_df['Gender_Inequality_Index_2014'], y = gender_df['Total_Unmet_Need'])
plt.ylabel('Index of Total Unmet Need for Family Planning')
plt.xlabel('Gender Inequality Index')
#plt.title('Gender Inequality and Family Planning')

plt.subplot(2,2,2)
plt.hist(x = gender_df['Male_Labor_force_2013'],color='blue', alpha=0.5, range=[0,100])
plt.hist(x = gender_df['Female_Labor_Force_2013'],color='yellow',alpha=0.5, range=[0,100])
plt.ylabel('# of Countries')
plt.xlabel('% Population Completing At Least Some Secondary School')
plt.title('Discrepancy in Education by Gender')
red_patch = mpatches.Patch(color='yellow', label='Females')
blue_patch = mpatches.Patch(color='blue', label='Males')
plt.legend(handles=[red_patch,blue_patch])
plt.show()

plt.subplot(2,2,3)
x = gender_df.as_matrix(columns=['Maternal_Mortality_2013'])
x= x.astype(np.int)     

x = np.array(x.T)
plt.ylabel('Maternal Mortality by Country 2013')
plt.title('Maternal Mortality Worldwide')

plt.boxplot(x)

plt.subplot(2,2,4)

plt.plot(gender_df['Gender_Inequality_Rank_2014'], gender_df['Gender_Inequality_Index_2014'])
plt.title('Country Assigned Gender Inequality Rank versus Index')
plt.ylabel('Gender Inequality Index')
plt.xlabel('Gender Inequality Rank')

plt.tight_layout()
plt.show()
```
